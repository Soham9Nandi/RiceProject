### Train Test Split

```python
#Define Features(Indedpendent Variables)
x = dataset.iloc[:,1:-1].values
```

```python
#Define target
y = dataset.iloc[:, -1].values
```

```python
#Importing the function to split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#Importing the function to standardize the training and testing data for smoother execution
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
#Creating a Random Forest Classifier
classifier_rf = RandomForestClassifier(random_state = 0)
classifier_rf.fit(x_train, y_train)
```

```python
#Predicting the target(y) with accordance to the testing data
y_pred = classifier_rf.predict(x_test)
#Predicting the probability for the Roc_Auc Score
y_prob = classifier_rf.predict_proba(x_test)
#Applying various metrics to check the validity of the model
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score,confusion_matrix, roc_auc_score
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average = 'micro')
prec = precision_score(y_test, y_pred, average='micro')
rec = recall_score(y_test, y_pred,average='micro')
roc_auc = roc_auc_score(y_test,y_prob,multi_class='ovr',average = 'micro')
```

```python
#Creating a dataframe to store measured validity of the model
results = pd.DataFrame([["Random Forest", acc, f1, prec, rec,roc_auc]],
                       columns = ['Model','Accuracy','F1 Score', 'Precision','Recall','ROC/AUC'])
results
```

```python
#Creating a confusion matrix to visualise the prediction pattern
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot = True)
```

```python
#Cross validating the model
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier_rf, X= x_train, y=y_train, cv = 10)
print("Accuracy is {:.2f}%".format(accuracies.mean()*100))
print("Standard Devation is {:.2f}%".format(accuracies.std()*100))
```
